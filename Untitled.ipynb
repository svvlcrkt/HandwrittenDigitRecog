```python
import cv2
import array
from skimage import io
import tensorflow as tf
from skimage import color
import matplotlib.pyplot as plt
from keras.preprocessing import image
from keras import datasets, layers, models
from keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
# Normalize pixel values to be between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0
#x_train -> training values, y_train -> digital values for training x
#x_test -> testing values , y_test -> digital values for testing x
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python
import numpy as np
# for x in range(0,60000):
plt.imshow(x_train[7777], cmap = "Greys")
```

```python
# displaying 1st 9 elements
plt.figure(figsize=(10,10))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.grid(False)
    plt.imshow(x_train[i],cmap = "Greys")
plt.show()
```

```python
model = models.Sequential()
model.add(layers.Conv2D(16, (5, 5), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (5, 5), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(64))
model.add(layers.Dense(10))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
#fitting the model
history = model.fit(x_train,y_train,epochs=5,batch_size=1,verbose=1)
```

```python
import matplotlib.pyplot as plt

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
model.evaluate(x_test, y_test,batch_size=1)
```

```python
def ret_x_cord_contour(contours):
    if(cv2.contourArea(contours) > 50):
        cent_moment = cv2.moments(contours)
        return (int(cent_moment['m10']/cent_moment['m00']))
    else:
        return 0
```

```python
# Find contours, obtain bounding box, extract and save ROI
import array
images = []

cnts, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
sorted_contours = sorted(cnts, key = ret_x_cord_contour, reverse=True)
for c in sorted_contours:
    i=0
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(thresh, (x, y), (x + w, y + h), (0, 0 ,0), 0)
    ROI = thresh[y:y+h, x:x+w]
    if (ROI.shape[1]*ROI.shape[0])>50:
        images.insert(i, ROI)
        i += 1
plt.imshow(thresh,cmap="Greys")    
cv2.waitKey()
```

```python
for i in range(len(images)):
    if (round(images[i].shape[0]/images[i].shape[1]) | round(images[i].shape[1]/images[i].shape[0])) < 5:
        images[i] = cv2.resize(images[i], (images[i].shape[0], round((images[i].shape[0])*1.5)),interpolation=cv2.INTER_AREA)
```

```python
for i in range(len(images)):
    plt.imshow(images[i], cmap='Greys')
    plt.title(f'Image {i+1}')
    plt.show()
```

```python
# Initialize an empty list to store the result images
result_images = []
for i in range(len(images)):
    old_image_height, old_image_width = images[i].shape
    new_image_width = round((images[i].shape[0]) * 1.75)
    new_image_height = round((images[i].shape[0]) * 1.75)
    clr = (0)
    result = np.full((new_image_height,new_image_width), clr, dtype=np.uint8)
    x_center = int((new_image_width - old_image_width) / 2)
    y_center = int((new_image_height - old_image_height) / 2)
    result[y_center:y_center+old_image_height, 
           x_center:x_center+old_image_width] = images[i]
    
    # Append the result image to the list
    result_images.append(result)
    cv2.destroyAllWindows()

```

```python
for i in range(len(images)):
    plt.imshow(images[i], cmap='Greys')
    plt.title(f'Image {i+1}')
    plt.show()
```

```python
# Create a list to store the dilated images
dilated_images = []

# Iterate over each image in `result_images` and apply dilation
for img in result_images:
    dilated_img = cv2.dilate(img, kernel, iterations=1)
    dilated_images.append(dilated_img)
```

```python
for i in range(len(dilated_images)):
    plt.imshow(dilated_images[i], cmap='Greys')
    plt.title(f'Image {i+1}')
    plt.show()
```

```python
# Create a list to store the dilated images
results = []
for img in dilated_images:
    result3 = cv2.resize(img, (28, 28) , interpolation = cv2.INTER_AREA)
    result3 = result3.astype('float64')
    result3 /= 255.0
    results.append(result3)
```
